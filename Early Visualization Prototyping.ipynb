### Scatterplots

```python
plt.figure(figsize=(8, 6))

years = ["2022", "2023", "2024"]
colors = ["darkblue", "seagreen", "sienna"]
markers = ["o", "s", "^"]

for year, color, marker in zip(years, colors, markers):
    sns.scatterplot(
        x=df[f"{year} Passing Touchdowns"],
        y=df[f"{year} PPR Fantasy Points Scored"],
        label=year,
        color=color,
        marker=marker
    )

plt.title("Overlayed Passing Touchdowns vs PPR Fantasy Points")
plt.legend(title="Year")
plt.show()
```

### Correlation Matrix

```python
corr_2022 = df[['2022 Age', '2022 Games Played',
       '2022 Games Started', '2022 Passing Attempts',
       '2022 Passing Completions', '2022 Passing Yards',
       '2022 Passing Touchdowns', '2022 Interceptions Thrown',
       '2022 Rushing Attempts', '2022 Rushing Yards',
       '2022 Rushing Touchdowns', '2022 Targets', '2022 Receptions',
       '2022 Receiving Yards', '2022 Receiving Touchdowns', '2022 Fumbles',
       '2022 Fumbles Lost', '2022 ESPN ADP', '2022 Sleeper ADP', '2022 NFL ADP',
       '2022 RTSports ADP', '2022 Average ADP', '2022 Positional ADP',
       '2022 PPR Fantasy Points Scored',
       '2022 Standard Fantasy Points Scored']].corr()

plt.figure(figsize=(24, 18))
sns.heatmap(corr_2022, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
